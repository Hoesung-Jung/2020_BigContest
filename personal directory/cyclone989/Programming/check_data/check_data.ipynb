```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("2020빅콘테스트_스포츠투아이_제공데이터_팀타자_2016.csv")
```

```python
G = 0
H = 0
for i in range(len(data['T_ID'])):
    if data['T_ID'][i] == 'HH':
        G += 1
        H += data['HIT'][i]
    if G == 113:
        break

print(H)
```

```python
def check_data(data, cols, team='HH'):
    col_vals = []
    for col in cols:
        col_val = 0
        G = 0
        for i in range(len(data['T_ID'])):
            if data['T_ID'][i] == team:
                G += 1
                col_val += data[col][i]
            if G == 113:
                col_vals.append(col_val)
                break
                
    return col_vals
```

```python
cols = ['HIT', 'H2', 'H3', 'HR']
```

```python
check_vals = check_data(data, cols)
```
