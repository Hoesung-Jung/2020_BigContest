# OPS = 장타율 + 출루율

#### 출루율(OBP) = (H + BB + HBP) / (AB + BB + HBP + SF)

#### 장타율(SLG) = (H + 2B + 3B * 2 + 3 * HR) /  AB

# WHIP = (HA + BBA) / IP

```python
data_full = pd.read_csv("../../../datasets/2016to2020_example_data/kbo_data_2016to2020.csv")
data_full
```

```python
OBP = []
for i in range(len(data_full['H'])):
    H = data_full['H'][i]
    BB = data_full['BB'][i]
    HBP = data_full['HBP_H'][i]
    AB = data_full['AB'][i]
    SF = data_full['SF_H'][i]
    OBP.append((H + BB + HBP) / (AB + BB + HBP + SF))
```

```python
SLG = []
for i in range(len(data_full['H'])):
    H = data_full['H'][i]
    B2 = data_full['2B'][i]
    B3 = data_full['3B'][i]
    HR = data_full['HR'][i]
    AB = data_full['AB'][i]
    SLG.append((H + B2 + B3 * 2 + 3 * HR) / AB)
```

```python
OBP = np.array(OBP)
OBP = OBP.reshape(1, 50)

SLG = np.array(SLG)
SLG = SLG.reshape(1, 50)

OPS = OBP + SLG
OPS = OPS.reshape(50, )
OPS = list(OPS)
```

```python
data_full['OPS'] = OPS
```

```python
WHIP = []
for i in range(len(data_full['HA'])):
    HA = data_full['HA'][i]
    BBA = data_full['BBA'][i]
    IP = data_full['IPOuts'][i] / 3
    WHIP.append((HA + BBA) / IP)
```

```python
data_full['WHIP'] = WHIP
```
