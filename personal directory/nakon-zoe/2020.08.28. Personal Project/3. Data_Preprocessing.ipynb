```python
import pandas as pd
```

# Make 2019 Data

### Make Pre_Data

```python
df19_pre = pd.read_csv('3. SplitedDatasets/2019_pre.csv')
df19_pre
```

```python
df19_pre = df19_pre.drop(['yearID', 'HEADER_NO', 'gameNO', 'GDAY_DS', 'VS_T_ID', 'TB_SC', 'WLS'], axis='columns')
df19_pre
```

```python
teamlist=set()
for item in df19_pre['T_ID']:
    teamlist.add(item)
teamlist=list(teamlist)
teamlist
```

```python
df19 = pd.DataFrame(df19_pre)
df19 = df19.drop([i for i in range(len(df19))])
df19
```

```python
for i in range(len(teamlist)):
    temp = pd.DataFrame(df19_pre[df19_pre['T_ID']==teamlist[i]].sum()).T
    df19 = pd.concat([df19, temp])
del df19['T_ID']
df19
```

### Make Post_Data

```python
df19_post = pd.read_csv('3. SplitedDatasets/2019_post.csv')
df19_post
```

```python
del_list=[]
for item in list(df19_post.columns):
    if not (item=='WLS' or item=='T_ID'):
        del_list.append(item)
df19_post = df19_post.drop(del_list, axis='columns')
df19_post
```

```python
for i in range(len(df19_post)):
    if df19_post['WLS'][i]=='W':
        df19_post['WLS'][i]=1
    elif df19_post['WLS'][i]=='L' or df19_post['WLS'][i]=='D':
        df19_post['WLS'][i]=0
df19_post
```

```python
df19_temp = pd.DataFrame(df19_post)
df19_temp = df19_post.drop([i for i in range(len(df19_post))])
df19_temp
```

```python
for i in range(len(teamlist)):
    temp = pd.DataFrame(df19_post[df19_post['T_ID']==teamlist[i]].sum()).T
    df19_temp = pd.concat([df19_temp, temp], sort=True)
del df19_temp['T_ID']
df19_temp
```

### Make 2019_Data

```python
df19['Post_WLS'] = df19_temp['WLS']
df19
```
