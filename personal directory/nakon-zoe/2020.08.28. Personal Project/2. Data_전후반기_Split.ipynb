```python
import pandas as pd
```

# 2019 Split

```python
df = pd.read_excel('3. SplitedDatasets/Original/2019_TotalData.xlsx')
df
```

```python
del_list_post=[]; del_list_pre=[];
for i in range(len(df)):
    if df['gameNO'][i]>113:
        del_list_post.append(i)
    else:
        del_list_pre.append(i)
```

```python
df_pre = df.drop(del_list_post)
df_pre
```

```python
df_post = df.drop(del_list_pre)
df_post
```

```python
df_pre.to_csv('3. SplitedDatasets/2019_pre.csv', index = False, encoding = 'cp949')
df_post.to_csv('3. SplitedDatasets/2019_post.csv', index = False, encoding = 'cp949')
```
