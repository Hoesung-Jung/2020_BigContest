```python
import pandas as pd
import numpy as np
```

```python
data_team_pitcher = pd.read_csv('2020빅콘테스트_스포츠투아이_제공데이터_팀투수_2016.csv')
data_team_batter = pd.read_csv('2020빅콘테스트_스포츠투아이_제공데이터_팀타자_2016.csv')
```

# WSWin; 한국시리즈 우승

```python
dictout = {}
for i in range(len(data_team_pitcher['T_ID'])):
    temp = data_team_pitcher['T_ID'][i]
    if temp not in dictout:
        dictout[temp] = []
        if data_team_pitcher['T_ID'][i] == 'OB':
            dictout[temp].append('1')
        else:
            dictout[temp].append('0')
```

```python
val = list(dictout.values())
val
#range(len(val))
```

```python
arr = []
for num in range(len(val)):
    arr.append(int(val[num][0]))
arr
```

# R; 득점

```python
data_team_batter
```

```python
r_dict = {}
for i in range(len(data_team_batter['T_ID'])):
    temp = data_team_batter['T_ID'][i]
    if temp not in r_dict:
        r_dict[temp] = data_team_batter['RUN'][i]
    else:
        r_dict[temp] += data_team_batter['RUN'][i]
r_dict
```

```python
AB_dict = {}
for i in range(len(data_team_batter['T_ID'])):
    temp = data_team_batter['T_ID'][i]
    if temp not in AB_dict:
        AB_dict[temp] = data_team_batter['AB'][i]
    else:
        AB_dict[temp] += data_team_batter['AB'][i]
AB_dict
```

```python
HIT_dict = {}
for i in range(len(data_team_batter['T_ID'])):
    temp = data_team_batter['T_ID'][i]
    if temp not in HIT_dict:
        HIT_dict[temp] = data_team_batter['HIT'][i]
    else:
        HIT_dict[temp] += data_team_batter['HIT'][i]
HIT_dict
```

```python
H2_dict = {}
for i in range(len(data_team_batter['T_ID'])):
    temp = data_team_batter['T_ID'][i]
    if temp not in H2_dict:
        H2_dict[temp] = data_team_batter['H2'][i]
    else:
        H2_dict[temp] += data_team_batter['H2'][i]
H2_dict
```

```python
H3_dict = {}
for i in range(len(data_team_batter['T_ID'])):
    temp = data_team_batter['T_ID'][i]
    if temp not in H3_dict:
        H3_dict[temp] = data_team_batter['H3'][i]
    else:
        H3_dict[temp] += data_team_batter['H3'][i]
H3_dict
```

```python
HR_dict = {}
for i in range(len(data_team_batter['T_ID'])):
    temp = data_team_batter['T_ID'][i]
    if temp not in HR_dict:
        HR_dict[temp] = data_team_batter['HR'][i]
    else:
        HR_dict[temp] += data_team_batter['HR'][i]
HR_dict
```

```python
BB_dict = {}
for i in range(len(data_team_pitcher['T_ID'])):
    temp = data_team_pitcher['T_ID'][i]
    if temp not in BB_dict:
        BB_dict[temp] = data_team_pitcher['BB'][i]
    else:
        BB_dict[temp] += data_team_pitcher['BB'][i]
BB_dict
```

```python
KK_dict = {}
for i in range(len(data_team_pitcher['T_ID'])):
    temp = data_team_pitcher['T_ID'][i]
    if temp not in KK_dict:
        KK_dict[temp] = data_team_pitcher['KK'][i]
    else:
        KK_dict[temp] += data_team_pitcher['KK'][i]
KK_dict
```

```python
year_dict = {}
for i in range(len(data_team_batter['T_ID'])):
    temp = data_team_pitcher['T_ID'][i]
    if temp not in year_dict:
        year_dict[temp] = 2016
year_dict
```

```python
# 2020까지 저장 완료
wswin = []
for num in range(len(val)):
    wswin.append(int(val[num][0]))
df = pd.DataFrame({'year' : list(year_dict.values()),
                   'team' : list(dictout.keys()),
                  'WSWin' : wswin,
                  'R' : list(r_dict.values()),
                  'AB' : list(AB_dict.values()),
                  'H' : list(HIT_dict.values()),
                  '2B' : list(H2_dict.values()),
                  '3B' : list(H3_dict.values()),
                  'HR' : list(HR_dict.values()),
                  'BB' : list(BB_dict.values()),
                  'SO' : list(KK_dict.values())})
df.to_csv('2016.csv', index = False, encoding = 'cp949')
```
