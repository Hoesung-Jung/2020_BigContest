타율데이터 생성

```python
def avg(df):
    hit_all = df['안타']
    avg_day = hit_all/df['타수']
    df['타율'] = avg_day
    df['득점권타율'] = df['득점권안타']/df['득점권타수']
```

시계열 데이터를 위해 누적데이터를 만들어낸다.

```python
def accum_bat(df):
    df_mod = df.loc[:,'타자':]
    df_accum = []
    col = df.columns[6:]
    for i in range(len(df_mod)):
        if i == 0:
            df_accum.append(df_mod.values[i])
        else:
            df_accum.append(df_accum[i-1] + df_mod.values[i])
    df_accum = pd.DataFrame(df_accum, columns=col)
    
    #일자 컬럼 추가
    ser_date = pd.to_datetime(df['일자'], format='%Y%m%d',errors='ignore')
    ser_date_mod = ser_date.reset_index()
    df_accum['일자']=ser_date_mod['일자']
    
    #타율 드랍
    df_accum = df_accum.drop('득점권타율', axis=1)
    return df_accum
```

이제 예열은 끝났다. 팀별 코드를 만들고 위 함수를 수정해보자.

- accum_bat함수 수정:  1.return의 dataframe에 date 컬럼을 추가하였다. 2. 득점권타율 column 삭제. 
- accum_bat함수 수정:  3. datetime이 NaT로 사라지는 현상발견: 이는 to_datetime에 시리즈를 넣었기에 row값이 저장된 상태로 들어갔다. 다시 인덱싱 필요함.

```python
def team_sep(df,team):
    df_new =df[df['팀코드']==team]
    return df_new
```

```python
def per_fit(df):
    #타자부터 득점권안타까지 경기당 기록으로 change
    for i in range(len(df)):
        df.loc[i,:'득점권안타'] = df.loc[i,:'득점권안타']/(i+1)
```

타율을 label로 설정

```python
TEST_SIZE = 144
```

```python
train = df[:-TEST_SIZE]
test = df[-TEST_SIZE:]
```

```python
# 20일(window_size=20) 데이터를 묶어서 처리하는 함수
def make_dateset(data, label, window_size=10):
    import numpy as np
    feature_list = []
    label_list = []
    for i in range(len(data) - window_size):
        feature_list.append(np.array(data.iloc[i:i+window_size]))
        label_list.append(np.array(label.iloc[i+window_size]))
    return np.array(feature_list), np.array(label_list)
```

```python
feature_cols = df.columns[:-3]
label_cols = ['타율']
```

```python
train_feature, train_label = make_dateset(train[feature_cols], train[label_cols])
test_feature, test_label = make_dateset(test[feature_cols], test[label_cols])
```

```python
test_feature, test_label = make_dateset(test[feature_cols], test[label_cols])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(
    train_feature, train_label, test_size=0.1)
```

```python
x_train.shape, x_valid.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.layers import LSTM
```

```python
model = Sequential()
```

```python
model.add(LSTM(16, activation='relu',
              input_shape=(train_feature.shape[1], train_feature.shape[2])))
model.add(Dense(1))
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['mae'])
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=10)
```

```python
hist = model.fit(x_train, y_train, epochs=200, batch_size=16,
                validation_data=(x_valid, y_valid))
```

```python
pred = model.predict(test_feature)
```

```python
model.evaluate(test_feature, test_label)
```
